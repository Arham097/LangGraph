## _Persistance_

#### Persistance in LangGraph refers to the ability to save and restore state of a workflow over time

1. Persistance helps to implement Fault Tolerance (Means During Execution if on some node program crash so it can resume from the same node)

2. Can Resume States were you leave after some time (e.g Resume Converstaion with Chatbot which you did 2-3 days ago)


## Checkpointers in Persistance

How Checkpointers helps in persistance?

Checkpointers implements checkpoints in all the the graph, where each superstep becomes the check point, and when the execution of graph start, so checkpointers save the State values when it find the checkpoint, so in conclusion the state of graph (workflow) saved in memory/Database with respect to checkpoints not all in one go of whole graph at once.


## Threads in Persistance

Threads in Perstance helps to made distinction/sepration of results of workflows.
For Example:
A workflow executed with initial input (5) and ater all calculation it gives result (total = 2)
Again the same workflow executed with initial input (6) and ater all calculation it gives result (total = 3)

So if i want to get the result of any one of the above workflow, so how do i get it ? means how the same workflow with different results fetch properly as we want.

So the answer is threads, when executing graph a _thread Id_ is provided, for example: for workflow 1, thread id is 1
and for workflow 2, thread id 2, so we can fetch the results/states of workflow by providing thread id specific to that workflow

1. Short Term Memory
2. Fault Tolerance
